```
import re 
def preproc(text):
    text = re.sub(" +", " ", text)
    useless_symblos = ["\\n","\",","\\","-","—","\'","\"", "«","»","(", ")","&quot"]
    #useless_symblos = ["\\n","«","»","&quot"]
    for sym in useless_symblos:
        text = text.replace(sym," ")
    text = re.sub(" +", " ", text)
    text = text.replace(" .",". ")
    text = text.replace(" ,",", ")
    text = text.replace(" ?","? ")
    text = text.replace(" !","! ")
    text = text.replace(" :",": ")
    text = text.replace(" ;","; ")
    text = re.sub(" +", " ", text)
    text = text.strip()

    return(text)
```

```
#will be usefull to find simularity between sentence
def cos(vector1,vector2):
    return np.dot(np.array(vector1),np.array(vector2))/(((np.sum(vector1**2))**0.5) * ((np.sum(vector2**2))**0.5))
```

```
#Find mean of every sentence cased on vectors from fasttext
#Find mean of this vectors (May be it collects the meaning of arteicle)
#Get cosie distance betweer the mean vector of text and all vectors of sentence
#Sort them and get the first 300 chars for comparison 
import numpy as np 
from nltk.tokenize import sent_tokenize


#This variable will contain the referat
homework6 = []
#number of observation
l = len(dataset)
for ind in range(l):
    
    text = dataset[ind] 
    sentences = sent_tokenize(text)
    mean_of_sentence = []
    mean_of_text = []
    for_mean = []
    for sentence in sentences:
        vec_of_word = []
        
        tokens = sentence.split(" ")
        for token in tokens:
          try:
            vec_of_word.append(fasttext.wv[token])
          except: 
            continue
        yyy = np.mean(vec_of_word, axis = 0)
        mean_of_sentence.append([sentence, yyy])
        for_mean.append(yyy)
    mean_of_text = np.mean(for_mean, axis = 0)
    cos_dist= [] 
    for k in range(len(mean_of_sentence)):
        cos_dist.append([mean_of_sentence[k][0], cos(mean_of_sentence[k][1],mean_of_text)])
    sort = sorted(cos_dist, key=lambda x: x[1], reverse=True)
    text_of_abs = ''
    for j in range(len(sort)):
        sentence = sort[j][0]
        text_of_abs = text_of_abs + sentence +" "
    token1 = text_of_abs.split(" ")
    referat = ''
    for token in token1: 
        if (len(referat) < 300):
            referat = referat + token + " "
           
    #in some cases len of referat can be longer than 300 because it is compared after concatination
    #so we fave to delete the last word
    if (len(referat) > 300):
        referat1 = '' 
        tokens_del = referat.split(" ")
        #2 because the last token always is empty ('')
        tokens_safe = tokens_del[:(len(tokens_del)-2)]
        for token in tokens_safe:
            referat1 = referat1 + token + " "
    else:
        referat1 = referat
    text_of_abs = preproc(referat1)
    #print('Abstract: ',text_of_abs)
    homework6.append(text_of_abs)
```
